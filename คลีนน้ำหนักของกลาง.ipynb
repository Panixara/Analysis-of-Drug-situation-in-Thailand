```python
import pandas as pd
from google.colab import files
read = pd.read_excel("/content/น้ำหนักของกลางยาเสพติด จำแนกตามชนิดยาเสพติด เป็นรายจังหวัด พ.ศ. 2554 - 2564.xlsx", header=2)
read
```

```python
read['จังหวัด'] = read['จังหวัด'].ffill()
read.info()
```

##แก้เมทแอมเฟตามีนเม็ดเป็นกิโลกรัม

```python
read.columns = read.columns.astype(str)
```

```python
df = pd.DataFrame(read)

for index, row in df.iterrows():
    if row['ชนิดยาเสพติด'] == 'เมทแอมเฟตามีน (เม็ด)':
        # แปลงเป็นกิโลกรัม
        df.at[index, '2561'] = row['2561'] / 10000
        df.at[index, '2562'] = row['2562'] / 10000
        df.at[index, '2564'] = row['2564'] / 10000

df
```

```python
# เลือกเฉพาะแถวที่ชนิดยาเสพติดเป็นเมทแอมเฟตามีน (เม็ด)
df_filtered = df[df['ชนิดยาเสพติด'] == 'เมทแอมเฟตามีน (เม็ด)']
df_filtered
```

```python
read.update(df_filtered)
```

```python
read[read['จังหวัด'] == 'กรุงเทพมหานคร']
```

##แก้เอ็กซ์ตาซีเม็ด เป็นกิโลกรัม

```python
df2 = pd.DataFrame(read)

for index, row in df.iterrows():
    if row['ชนิดยาเสพติด'] == 'เอ็กซ์ตาซี (เม็ด)':
        # แปลงเป็นกิโลกรัม
        df2.at[index, '2561'] = row['2561'] / 10000
        df2.at[index, '2562'] = row['2562'] / 10000
        df2.at[index, '2564'] = row['2564'] / 10000

df2
```

```python
# เลือกเฉพาะแถวที่ชนิดยาเสพติดเป็นเมทแอมเฟตามีน (เม็ด)
df_filtered2 = df2[df2['ชนิดยาเสพติด'] == 'เอ็กซ์ตาซี (เม็ด)'].round(2)

df_filtered2
```

```python
read.update(df_filtered2)
```

##นำแถวเม็ดไปแทนที่ด้านบน

```python
# แทนที่ค่า NaN ในแถวของ "เมทแอมเฟตามีน" โดยใช้ค่าใน "เมทแอมเฟตามีน (เม็ด)"
read.loc[read['ชนิดยาเสพติด'] == 'เอ็กซ์ตาซี', ['2561', '2562', '2564']] = read.loc[df['ชนิดยาเสพติด'] == 'เอ็กซ์ตาซี (เม็ด)', ['2561', '2562', '2564']].values
read.loc[read['ชนิดยาเสพติด'] == 'เมทแอมเฟตามีน', ['2561', '2562', '2564']] = read.loc[df['ชนิดยาเสพติด'] == 'เมทแอมเฟตามีน (เม็ด)', ['2561', '2562', '2564']].values

read
```

```python
read[read['จังหวัด'] == 'สมุทรปราการ']
```

```python
# เงื่อนไขสำหรับลบแถวที่มีชนิดยาเสพติดเป็นเมทแอมเฟตามีน (เม็ด)หรือเอ็กซ์ตาซี (เม็ด)
condition = read['ชนิดยาเสพติด'].isin(['เมทแอมเฟตามีน (เม็ด)', 'เอ็กซ์ตาซี (เม็ด)'])

read = read[~condition]
read
```
